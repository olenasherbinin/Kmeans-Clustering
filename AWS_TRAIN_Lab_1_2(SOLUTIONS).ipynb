#### **Problem #6:**

You are interested at looking at houses in Weston, MA. You want to come up with general statisitcs, so you look at some data from a house selling site. The data, `houses_df`, contains the location of each house as the town, state, and street. The associated statistics show the price and square footage of each house. Use the dataframe to answer Questions #7-9.

```python
house_dict = {"town":["Weston", "Weston", "Weston", "Weston", "Weston", "Weston", "Weston", "Weston", "Weston"],
              "state": ["MA", "MA", "MA", "MA", "MA", "MA", "MA", "MA", "MA"],
              "street": ["Webster", "Perry", "Loring", "Wellsley", "Lantern", "Jericho", "Lexington", "Colchester", "Partridge"],
              "price": [10400000, 9750000, 2895000, 5100000, 1790000, 799000, 1995000, 4250000, 2195000],
              "feet": [14056, 12795, 5778, 12148, 4663, 1611, 4600, 6700, 4070]
    
}
houses_df = pd.DataFrame(house_dict)
houses_df
```

#### **Problem #7:** 

Find the mean and median price and square feet of a house in Weston, MA.

##### **Solution**

```python
print(f"Mean Price, {houses_df['price'].mean()}, Median Price {houses_df['price'].median()}")
print(f"Mean sqft, {houses_df['feet'].mean()}, Median sqft {houses_df['feet'].median()}")
```

#### **Problem #8:** 

Find the price of the 6th house.

##### **Solution**

#### **Problem #9:** 

Iterate through the dataframe and create a list of the mean price per square foot. Use this list to calculate the average price per square foot of a house in Weston, MA.

##### **Solution**

```python
price_per_sqft = []

for i in range(len(houses_df.iloc[:,1])):
  price_per_sqft.append(houses_df.iloc[i,3] / houses_df.iloc[i,4])

sum(price_per_sqft)/len(houses_df.iloc[:,1])
```
